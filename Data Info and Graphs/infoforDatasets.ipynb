#### Dataset 1 (18700 entries)

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Dataset1.csv')

# Drop the 'Vehicle' column
df.drop('Vehicle ', axis=1, inplace=True)

# Calculate mean, median, and mode for each column
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]  # mode() returns a DataFrame, take the first row

print("Mean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nMode Values:")
print(mode_values)
```

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
df = pd.read_csv('Dataset1.csv')

# Drop the 'Vehicle' column
df.drop('Vehicle ', axis=1, inplace=True)

# Calculate summary statistics
summary_statistics = df.describe()

# Calculate mode (pandas mode() might return multiple rows if there is a tie, take the first row)
mode_values = df.mode().iloc[0]

# Add the mode to the summary statistics
summary_statistics.loc['mode'] = mode_values

summary_statistics
```

```python
import scipy.stats as stats

mean = 54.079659
std_dev = 11.964752

# Z-scores for 30th and 70th percentiles
z_30th = stats.norm.ppf(0.3)  # Percent point function (inverse of cdf)
z_70th = stats.norm.ppf(0.7)

# Calculate x and y
x = mean + z_30th * std_dev
y = mean + z_70th * std_dev

x, y
```

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
df = pd.read_csv('Dataset1.csv')

# Drop the 'Vehicle' column
df.drop('Vehicle ', axis=1, inplace=True)

# Calculate summary statistics
summary_statistics = df.describe()

# Calculate mode (pandas mode() might return multiple rows if there is a tie, take the first row)
mode_values = df.mode().iloc[0]

# Add the mode to the summary statistics
summary_statistics.loc['mode'] = mode_values

# Print the table using tabulate for better formatting
print(tabulate(summary_statistics, headers='keys', tablefmt='pretty'))
```
