## We will go ahead and create our Vehiclular Dataset from Dataset.csv containing 1000 Vehicles

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('Dataset1.csv')

# Randomly select 500 vehicles from the original dataset
random_500 = df.sample(500, random_state=42)

# Select 900 vehicles where the eligibility score is under 25% which is 45.800069447
low_eligibility_500 = df[df['Eligible'] < 45.800069447].sample(500, random_state=42)

# Combine the two datasets
new_dataset = pd.concat([random_500, low_eligibility_500], ignore_index=True)

# Remove old Vehicle numbers and add new ones from 1 to 1000
new_dataset.reset_index(drop=True, inplace=True)
new_dataset.index += 1
new_dataset.index.name = 'Vehicle'

# Save the new dataset
new_dataset.to_csv('1000VehicleDataset.csv')

new_dataset.head()
```

## Creating our Datasets to Train the Machine Learning Models (contains 4000 entries)

- For the training dataset we randomly select 4000 vehicles from Dataset 2

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('Dataset2.csv')

# Randomly select 400 vehicles from the original dataset
random_10 = df.sample(4000, random_state=42)
new_dataset = random_10


# Check and drop any existing 'Vehicle' column if present
if 'Vehicle ' in new_dataset.columns:
    new_dataset.drop(columns='Vehicle ', inplace=True)
    
# Remove old Vehicle numbers and add new ones from 1 to 1000
new_dataset.reset_index(drop=True, inplace=True)
new_dataset.index += 1
new_dataset.index.name = 'Vehicle'

# Save the new dataset
new_dataset.to_csv('VehicleTrainingDataset.csv')

new_dataset.head()
```
