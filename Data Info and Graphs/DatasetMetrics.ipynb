# Calculating Metrics for VehicleTrainingDataset.csv

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
df = pd.read_csv('VehicleTrainingDataset.csv')

# Drop the 'Vehicle' column
df.drop('Vehicle', axis=1, inplace=True)

# Calculate summary statistics
summary_statistics = df.describe()

# Calculate mode (pandas mode() might return multiple rows if there is a tie, take the first row)
mode_values = df.mode().iloc[0]

# Add the mode to the summary statistics
summary_statistics.loc['mode'] = mode_values

summary_statistics
```
