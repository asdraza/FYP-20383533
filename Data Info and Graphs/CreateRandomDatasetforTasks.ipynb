100Tasks

```python
import pandas as pd

# Load datasets
df1 = pd.read_csv('dataset1.csv')
df2 = pd.read_csv('dataset2.csv')
df3 = pd.read_csv('dataset3.csv')

# Randomly sample 50 entries from dataset1
random_subset1 = df1.sample(n=50, random_state=42)

# Randomly sample 30 entries from dataset2
random_subset2 = df2.sample(n=30, random_state=42)

# Randomly sample 20 entries from dataset3
random_subset3 = df3.sample(n=20, random_state=42)

# Drop the "Vehicle" column and add "TaskID" column
random_subset1 = random_subset1.drop(columns=['Vehicle '])
random_subset2 = random_subset2.drop(columns=['Vehicle '])
random_subset3 = random_subset3.drop(columns=['Vehicle '])

# Add TaskID column ranging from 1 to 100
random_subset1['TaskID'] = range(1, 51)
random_subset2['TaskID'] = range(51, 81)
random_subset3['TaskID'] = range(81, 101)

# Rearrange columns to have "TaskID" as the first column
random_subset1 = random_subset1[['TaskID'] + [col for col in random_subset1.columns if col != 'TaskID']]
random_subset2 = random_subset2[['TaskID'] + [col for col in random_subset2.columns if col != 'TaskID']]
random_subset3 = random_subset3[['TaskID'] + [col for col in random_subset3.columns if col != 'TaskID']]

# Concatenate the subsets
final_subset = pd.concat([random_subset1, random_subset2, random_subset3], ignore_index=True)

# Save the final subset to a new CSV file
final_subset.to_csv('RandomTasks100.csv', index=False)
```
